## Importing Required Libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
import plotly.graph_objects as go
```

## For Bitcoin

```python
cg = CoinGeckoAPI() 
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=90)
bitcoin_data
# Price, Market_Caps, Total Volumes
```

```python
bitcoin_price_data = bitcoin_data['prices']
```

```python
dataBT = pd.DataFrame(bitcoin_price_data, columns=['Timestamp','Price'])
dataBT
```

```python
dataBT['Date'] = pd.to_datetime(dataBT['Timestamp'], unit='ms')
dataBT
```

## Candlestick Chart Visualization

```python
candlestick_dataBT = dataBT.groupby(dataBT.Date.dt.date).agg({'Price' : ['min', 'max', 'first', 'last']})
# candlestick_dataBT = candlestick_dataBT.set_index('Date')
candlestick_dataBT
```

```python
fig = go.Figure(data=[go.Candlestick(x=candlestick_dataBT.index, 
                                        open=candlestick_dataBT['Price']['first'], 
                                        high=candlestick_dataBT['Price']['max'], 
                                        low=candlestick_dataBT['Price']['min'], 
                                        close=candlestick_dataBT['Price']['last'])])
```

```python
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date', 
                  yaxis_title='Price (USD $)', title='Bitcoin Candlestick Chart Over Past 90 Days' )
```
